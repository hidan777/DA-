```python
import pandas as pd
 
import os

current_directory = os.getcwd()

# First, we focus on the US_OIL dataframe. We fill NA values with 0 and select important features.

df_US_OIL = pd.read_excel(current_directory+"\\US OIL Dataset.xlsx", sheet_name = 'Data 2')

df_US_OIL.columns = df_US_OIL.iloc[1]
df_US_OIL = df_US_OIL[2:].reset_index(drop = True)
df_US_OIL['Date'] = pd.to_datetime(df_US_OIL['Date']).dt.strftime('%Y-%m')
df_US_OIL = df_US_OIL.fillna(0)
df_US_OIL
```

```python
print(df_US_OIL.columns)
df_US_OIL_Selected = df_US_OIL[['Date','U.S. Field Production of Crude Oil (Thousand Barrels)',
                                'U.S. Imports of Crude Oil (Thousand Barrels)', 'U.S. Exports of Crude Oil (Thousand Barrels)',
                                'U.S. Ending Stocks of Crude Oil (Thousand Barrels)']]
```

```python
df_US_OIL_Selected
```

```python
# Second, for the inflation dataframe, we apply the melt method and create a new date with the format '2023-08'.

df_inflation = pd.read_excel(current_directory+"\\Inflation table.xlsx")

print(df_inflation)


df_inflation = df_inflation[df_inflation['Year'].notna()]
df_inflation.drop(columns=['Ave'],inplace =True)
df_inflation['Year'] = df_inflation['Year'].astype(int).astype(str)

df_inflation_melt = df_inflation.melt(id_vars='Year', var_name='Month', value_name='Value')
df_inflation_melt['Date'] = df_inflation_melt['Year'] + '-' + df_inflation_melt['Month']
df_inflation_melt.drop(columns=['Year', 'Month'],inplace = True)

df_inflation_melt['Date'] = pd.to_datetime(df_inflation_melt['Date']).dt.strftime('%Y-%m')
df_inflation_melt.dropna()
```

```python
# Then, for the DFF dataframe, we group by Date and calculate the average DFF per month.

df_DFF = pd.read_csv(current_directory+"\\DFF.csv")
df_DFF['DATE'] =  pd.to_datetime(df_DFF['DATE']).dt.strftime('%Y-%m')
df_DFF = df_DFF.groupby(['DATE'])['DFF'].mean().round(2).reset_index()
df_DFF
```

```python
# Finally, we merge all the dataframes.

df_merged = df_DFF.merge(df_inflation_melt, left_on = 'DATE', right_on = 'Date', how = 'left')
df_merged = df_merged.merge(df_US_OIL_Selected, on= 'Date', how = 'left')
df_merged = df_merged.dropna().drop(columns = ['DATE']).rename(columns = {'Value' : 'U.S. Inflation', 'DFF' : 'Federal Fund Effective Rate (DFF)'})
```

```python
df_merged.to_csv(current_directory+'\inflation DFF OIL.csv',index = False)
df_merged
```
