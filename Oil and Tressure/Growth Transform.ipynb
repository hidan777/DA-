```python
import pandas as pd
 
import os

current_directory = os.getcwd()


df = pd.read_csv(current_directory+'\inflation DFF OIL.csv')


df = df[df.Date >= '2004-12']



print(df.columns)
i = df.columns

change = {}  # create empty dict
for i in df.columns:
    if i != 'Date':
        month3 = ((df[i].iloc[-1] - df[i].iloc[-4]) * 100/df[i].iloc[-4]).round(2)m
        month6 = ((df[i].iloc[-1] - df[i].iloc[-7]) * 100/df[i].iloc[-7]).round(2)
        year1 = ((df[i].iloc[-1] - df[i].iloc[-13]) * 100/df[i].iloc[-13]).round(2)
        year3 = ((df[i].iloc[-1] - df[i].iloc[-37]) * 100/df[i].iloc[-37]).round(2)
        year5 = ((df[i].iloc[-1] - df[i].iloc[-61]) * 100/df[i].iloc[-61]).round(2)
        year10 = ((df[i].iloc[-1] - df[i].iloc[-121]) * 100/df[i].iloc[-121]).round(2)
        year15 = ((df[i].iloc[-1] - df[i].iloc[-181]) * 100/df[i].iloc[-181]).round(2)
        yearmax = ((df[i].iloc[-1] - df[i].iloc[1]) * 100/df[i].iloc[1]).round(2)
        change['Growth of '+i + '(%)'] = month3,month6,year1,year3,year5,year10,year15,yearmax # tambahkan hasil ke dalam kamus dengan kunci 'change'+i

print(change)  
```

```python
Growth = pd.DataFrame(change)
Growth['Time Period'] = ['3 months','6 months','1 year','3 years','5 years','10 years','15 years','Max year']
Growth['No'] = [1,2,3,4,5,6,7,8]
Growth.to_csv(current_directory+'\inflation DFF OIL Growth.csv',index = False)
Growth
```
