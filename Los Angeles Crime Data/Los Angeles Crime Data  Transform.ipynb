```python
import pandas as pd 
import os

current_directory = os.getcwd()

df = pd.read_csv(current_directory+"\Crime_Data_from_2020_to_Present.csv")
df.columns
```

```python
df
```

```python
#creating Daytime phase features & Grouping Victim’s Gender

df['Time'] = df['TIME OCC'].apply(lambda x: f'{str(x).zfill(4)[:-2]}:{str(x).zfill(4)[-2:]}')
df['Time'] = pd.to_datetime(df['Time']).dt.time

df['Daytime Phase'] = df['Time'].apply(lambda x : 'Dawn' if x < pd.to_datetime('05:00').time() else 'Morning' if x < pd.to_datetime('12:00').time()
                                         else 'Afternoon' if x < pd.to_datetime('17:00').time() else 'Evening' if x < pd.to_datetime('20:00').time()
                                         else 'Night')

df['Victim Sex'] = df['Vict Sex'].apply(lambda x : 'Female' if x == 'F' else 'Male' if x == 'M' else 'Others')
```

```python
#Convert features into datetime type
df['Date Rptd'] = pd.to_datetime(df['Date Rptd'])
df['DATE OCC'] = pd.to_datetime(df['DATE OCC'])
```

```python
#Select important features
df_selected = df[['DR_NO', 'DATE OCC','AREA NAME', 'Crm Cd', 'Crm Cd Desc',
                  'Vict Age', 'Victim Sex', 'Premis Cd', 'Premis Desc',
                  'Weapon Used Cd', 'Weapon Desc', 'Status', 'Status Desc', 'LAT',
                  'LON','Daytime Phase']]

df_selected.to_csv(current_directory+'\Crime_Data_from_2020_to_Present_Final.csv',index = False)
```

```python
# making Discrenpancy dataframe
df['Report-Event Discrepancy'] = df['Date Rptd'] - df['DATE OCC']
df['Year'] = df['DATE OCC'].dt.year

df_Discrenpancy = pd.DataFrame(df.groupby('Year')['Report-Event Discrepancy'].mean().reset_index())
df_Discrenpancy['Day'] = df_Discrenpancy['Report-Event Discrepancy'].dt.days
df_Discrenpancy.to_csv(current_directory+'\Crime_Data_from_2020_to_Present_Discrepancy.csv', index = False)
```
